```python
import cv2
import matplotlib.pyplot as plt

# Charger l'image
image_path = r"C:\Users\pierrontl\OneDrive - GIE SIMA\Documents\GitHub\Fraude\code_Tom\filtre_noir\FACTURE A CONTROLERV2.jpg"  # Remplacez par le chemin de votre image
image = cv2.imread(image_path)

# Convertir l'image en niveaux de gris
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Appliquer le filtre noir et blanc
(thresh, black_and_white_image) = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Afficher l'image originale et l'image en noir et blanc
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Image Originale')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Image Noir et Blanc')
plt.imshow(black_and_white_image, cmap='gray')

plt.show()
```
