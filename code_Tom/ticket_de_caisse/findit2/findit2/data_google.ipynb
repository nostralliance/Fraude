```python
import os
import requests
from bs4 import BeautifulSoup
import uuid

def scrap_img_tags(url, output_dir):
    # Définir les en-têtes d'utilisateur
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    # Faire une requête GET avec les en-têtes d'utilisateur
    response = requests.get(url, headers=headers)
    
    # Vérifier si la requête a réussi (code de statut HTTP 200)
    if response.status_code == 200:
        # Parser le contenu HTML avec BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Trouver toutes les balises <img>
        img_tags = soup.find_all('img')
        
        # Créer le dossier de sortie s'il n'existe pas déjà
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        
        # Enregistrer chaque image dans le dossier de sortie
        for img in img_tags:
            img_url = img['src']
            img_name = str(uuid.uuid4()) + '.jpg'  # Générer un nom de fichier unique
            img_path = os.path.join(output_dir, img_name)
            
            # Télécharger l'image et l'enregistrer dans le dossier de sortie
            with open(img_path, 'wb') as f:
                f.write(requests.get(img_url).content)
                print(f"Image enregistrée: {img_path}")
    else:
        print("La requête a échoué. Statut code:", response.status_code)

# URL de la page web à scraper
url = 'https://fr.freepik.com/photos-vecteurs-libre/ticket-caisse-vierge'

# Dossier de sortie pour enregistrer les images
output_directory = r'C:\Users\tomlo\Documents\GitHub\Fraude\code_Tom\ticket_de_caisse\findit2\findit2\images'

# Appeler la fonction pour scraper les balises <img>
scrap_img_tags(url, output_directory)
```
