```python
import fitz

def lire_pdf(nom_fichier):
    doc = fitz.open(nom_fichier)
    
    for page_num in range(doc.page_count):
        page = doc[page_num]
        text = page.get_text()
        print(f"Page {page_num + 1} :\n{text}\n")
    
    doc.close()

fichier_pdf = 'faux_test.pdf'
res = lire_pdf(fichier_pdf)
print(res)
```
