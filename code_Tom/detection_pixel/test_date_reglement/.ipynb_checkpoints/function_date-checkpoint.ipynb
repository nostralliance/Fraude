```python


pngText = "une première date 17/10/2023 et voici la deuxième 14/02/2023 : 15,45 euro"

# On récupère la liste des dates dans le texte
dateList = re.findall(r'([0-2]{1}[0-9]{1})[/-](1[0-2]{1}|0[1-9]{1})[/-]([0-9]{2,4})', pngText)
print("Dates trouvées :", dateList)

# On récupère la date de règlement
date_reglement_regle = re.findall(r'réglé le (0[1-9]|[12]\d|3[01])/(0[1-9]|1[0-2])/((?:\d{4}|\d{2}))', pngText)
date_reglement_destinataire = re.findall(r'(0[1-9]|[12]\d|3[01])/(0[1-9]|1[0-2])/((?:\d{4}|\d{2}))\sau destinataire', pngText)
date_reglement_euro = re.findall(r'(0[1-9]|[12]\d|3[01])/(0[1-9]|1[0-2])/((?:\d{4}|\d{2})) : \d{1,3},\d{1,2} euro', pngText)

print("date reglement regle :", date_reglement_regle)
print("date reglement destinataire :", date_reglement_destinataire)
print("date reglement euro :", date_reglement_euro)

# Ajouter des variables pour chaque détection
if date_reglement_regle:
    # Convertir la date de règlement en objet datetime
    date_reglement_regle = date(int(date_reglement_regle[0][2]), int(date_reglement_regle[0][1]), int(date_reglement_regle[0][0]))
    # Pour chaque date récupérée
    for dateSplit in dateList:
        dateFormat = date(int(dateSplit[2]), int(dateSplit[1]), int(dateSplit[0]))
        # Comparer si une des dates de dateList est supérieure à la date de règlement
        if dateFormat > date_reglement_regle:
            print("------Une date dans dateList est supérieure à la date de règlement ! (regle)")
        else:
            print("Aucune date n'est supérieure à la date de règlement")

elif date_reglement_destinataire:
    # Convertir la date de règlement en objet datetime
    date_reglement_destinataire = date(int(date_reglement_destinataire[0][2]), int(date_reglement_destinataire[0][1]), int(date_reglement_destinataire[0][0]))
    # Pour chaque date récupérée
    for dateSplit in dateList:
        dateFormat = date(int(dateSplit[2]), int(dateSplit[1]), int(dateSplit[0]))
        # Comparer si une des dates de dateList est supérieure à la date de règlement
        if dateFormat > date_reglement_destinataire:
            print("------Une date dans dateList est supérieure à la date de règlement ! (destinataire)")
        else:
            print("Aucune date n'est supérieure à la date de règlement")

elif date_reglement_euro:
    # Convertir la date de règlement en objet datetime
    date_reglement_euro = date(int(date_reglement_euro[0][2]), int(date_reglement_euro[0][1]), int(date_reglement_euro[0][0]))
    # Pour chaque date récupérée
    for dateSplit in dateList:
        dateFormat = date(int(dateSplit[2]), int(dateSplit[1]), int(dateSplit[0]))
        # Comparer si une des dates de dateList est supérieure à la date de règlement
        if dateFormat > date_reglement_euro:
            print("------Une date dans dateList est supérieure à la date de règlement ! (euro)")
        else:
            print("Aucune date n'est supérieure à la date de règlement")
else:
    print("Aucune date de règlement n'a été trouvée.")
```

```python
import fitz

def lire_pdf(nom_fichier):
    doc = fitz.open(nom_fichier)
    
    for page_num in range(doc.page_count):
        page = doc[page_num]
        text = page.get_text()
        print(f"Page {page_num + 1} :\n{text}\n")
    
    doc.close()

fichier_pdf = 'faux_test.pdf'
res = lire_pdf(fichier_pdf)
print(res)
```

```python
from datetime import datetime
import re


def compare(date_simple_str, date_reglement_str):
    # convertir objet str en datetime
    date_simple = datetime.strptime(date_simple_str, "%d/%m/%Y")
    date_reglement = datetime.strptime(date_reglement_str, "%d/%m/%Y")

    # Comparer les dates
    if date_simple > date_reglement:
        print(f"{date_simple_str} est supérieure à {date_reglement_str}")
        return True
    else:
        print(f"{date_simple_str} n'est pas supérieure à {date_reglement_str}")
        return False

def extract_reglement_date(value):
    regex_regle = r'réglé le (\d{1,2}/\d{1,2}/\d{4})'
    match = re.search(regex_regle, value)
    if match:
        return match.group(1)
    return None

def isDateSimple(value):
    regex_simple = r'([0-2]{1}[0-9]{1})[/-](1[0-2]{1}|0[1-9]{1})[/-]([0-9]{2,4})'
    return re.match(regex_simple, value)

text = ['2/2', 'CAssurance', 'Maladie', 'Agir ensemble, protéger chacun', 'assuré social', 'TROJANOWSKI', 'JACQUELINE', 'n" de Sécurité Sociale 2 54 05 99 109 057 64', 'Pour toutes vos démarches; utilisez', 'CPAM du VAUCLUSE', 'le compte ameli', '84043 AVIGNON CEDEX 9', "ou l'application ameli pour smartphone", 'MME   TROJANOWSKI JACQUELINE', '55 CHE DU MOURVEDRE', '84400 GARGAS', 'Voici le détail des versements vous concemant pour la période du 01/1', '1/2023 au 30/11/2023', "Le remboursement des sommes que vous avez avancées est d'ores", 'Ces informations ont été directement transmises', 'et déjà effectué sur votre compte.', 'complémentaire MUTUELLE DE', "par votre caisse d'assurance maladie à votre organisme", 'pour information', "L'EST En conséquence; vous n'avez pas besoin de lui envoyerocganlevee", 'dates', 'nature des prestations', 'montant', 'base du', 'montant', 'payë   rembours .', 'versé', 'pour JACQUELINE né(e) le 24/05/1954', 'maladie', 'réf 8010   2331019905591', '31/10/2023', 'VEH.SANIT. LEGER', '(VSL', 'franchise à retenir ( 2 FRT) "', '82,32', '82,32', '55 %', '45,28', 'réglé le 07/11/2023 au destinataire AMBULANCES', '4,00', 'BONNIEUX', '45,28 euro(s)', 'maladie', 'ref 8010   2331140000266', '20/06/2022', 'TAXI', '(TXI', 'PC )', 'franchise à retenir ( 2 FRT) f"', '461,67', '461,67', '65 %', '300,09', 'réglé le 08/11/2023 au destinataire EURL MONDIAL', '4,00', 'TAXI ; 300,09 euro(s)', '(1) Les partlclpatlons forfaltalres et tranchlses constatées sur', 'de solns', 'ou le seront sur vos prochains règlements.', 'ces actes sont dédultes du montant remboursé dans la partle < Vos', 'remboursements']

myBlocs = []
currentBloc = []

def date_compare(ocr_result):
    started = False
    for value in ocr_result:
        if not started:
            currentBloc = []
            started = True

        if isDateSimple(value):
            currentBloc.append(value)

        reglement_date = extract_reglement_date(value)
        if reglement_date:
            currentBloc.append(reglement_date)
            myBlocs.append(currentBloc)
            currentBloc = []

    for block in myBlocs:
        date_reglement = block[-1]
        date_normales = block[:-1]

        for date in date_normales:
            result = compare(date, date_reglement)
            if result:
                print("True")
            else:
                print("False")
date_compare(text)
```
