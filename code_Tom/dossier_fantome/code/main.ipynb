# Connaitre l'age moyen 

```python
# Définir la date courante
current_date = pd.to_datetime('2024-07-24')

# Calculer les âges en prenant en compte les anniversaires
ages = (current_date.year - df_unique['Date de Naissance'].dt.year - 
        ((current_date.month < df_unique['Date de Naissance'].dt.month) | # vérifie si le mois actuel est avant le mois de naissance. Si c'est vrai, cela signifie que la personne n'a pas encore eu son anniversaire cette année.
         ((current_date.month == df_unique['Date de Naissance'].dt.month) & (current_date.day < df['Date de Naissance'].dt.day)))).tolist() # vérifie si c'est le même mois mais avant le jour d'anniversaire. Si c'est vrai, cela signifie que la personne n'a pas encore eu son anniversaire ce mois-ci.

liste_ages = [age for age in ages if pd.notna(age)]

moyenne = sum(liste_ages)/len(liste_ages)

print(moyenne)
```

La moyenne d'age est de 61 ans 

# Connaitre l'ecart-type de l'age 

```python
# Définir la date courante
current_date = pd.to_datetime('2024-07-24')

# Calculer les âges en prenant en compte les anniversaires
ages = (current_date.year - df_unique['Date de Naissance'].dt.year - 
        ((current_date.month < df_unique['Date de Naissance'].dt.month) | # vérifie si le mois actuel est avant le mois de naissance. Si c'est vrai, cela signifie que la personne n'a pas encore eu son anniversaire cette année.
         ((current_date.month == df_unique['Date de Naissance'].dt.month) & (current_date.day < df['Date de Naissance'].dt.day)))).tolist() # vérifie si c'est le même mois mais avant le jour d'anniversaire. Si c'est vrai, cela signifie que la personne n'a pas encore eu son anniversaire ce mois-ci.

# Création d'un DataFrame temporaire pour calculer l'écart-type des âges
ages_df = pd.DataFrame(liste_ages, columns=['Ages'])

# Calculer l'écart-type des âges
ect = ages_df['Ages'].std()

print("Ecart-type des âges:", ect)
```

Nous pouvons remarquer que l'age de l'écart-type par rapport a la moyenne qui est de 61 ans est asser dissipée, cela veut dire que la plus part des ages se situe a 78ans (61+17) et 44ans (61-17)

# Schéma représentant la moyenne d'age et l'écart-type sur le dataset de MUTEST

```python
# Créer un histogramme des âges
plt.figure(figsize=(10, 6))
plt.hist(liste_ages, bins=10, edgecolor='black')
plt.axvline(moyenne, color='red', linestyle='dashed', linewidth=1)
plt.axvline(moyenne+16, color='green', linestyle = 'dashed', linewidth=1)
plt.axvline(moyenne-16, color='green', linestyle = 'dashed', linewidth=1)
plt.title('Distribution des âges')
plt.xlabel('Âge')
plt.ylabel('Nombre de personnes')
plt.legend(['Moyenne d\'âge', 'écart-type'])
plt.show()
```

# Schéma qui représente le nombre de personne en fonction de l'année

```python
# Création d'un DataFrame pour les données à visualiser
data = {
    'Année': [2022, 2023, 2024],
    'Nombre de personnes en attente': [nombre_en_attente_2022, nombre_en_attente_2023, nombre_en_attente_2024]
}
df_plot = pd.DataFrame(data)

# Créer le diagramme en bâton avec Seaborn
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Année', y='Nombre de personnes en attente', data=df_plot, palette='viridis', estimator="sum")
ax.bar_label(ax.containers[0], fontsize=10)
ax.bar_label(ax.containers[1], fontsize=10)
ax.bar_label(ax.containers[2], fontsize=10)

plt.title('Nombre de personnes en attente par année')
plt.xlabel('Année')
plt.ylabel('Nombre de personnes en attente')
plt.show()
```
