# Connaitre l'age moyen 

```python
# D√©finir la date courante
current_date = pd.to_datetime('2024-07-24')

# Calculer les √¢ges en prenant en compte les anniversaires
ages = (current_date.year - df_unique['Date de Naissance'].dt.year - 
        ((current_date.month < df_unique['Date de Naissance'].dt.month) | # v√©rifie si le mois actuel est avant le mois de naissance. Si c'est vrai, cela signifie que la personne n'a pas encore eu son anniversaire cette ann√©e.
         ((current_date.month == df_unique['Date de Naissance'].dt.month) & (current_date.day < df['Date de Naissance'].dt.day)))).tolist() # v√©rifie si c'est le m√™me mois mais avant le jour d'anniversaire. Si c'est vrai, cela signifie que la personne n'a pas encore eu son anniversaire ce mois-ci.

liste_ages = [age for age in ages if pd.notna(age)]

moyenne = sum(liste_ages)/len(liste_ages)

print(moyenne)
```

La moyenne d'age est de 61 ans 

# Connaitre l'ecart-type de l'age 

```python
# D√©finir la date courante
current_date = pd.to_datetime('2024-07-24')

# Calculer les √¢ges en prenant en compte les anniversaires
ages = (current_date.year - df_unique['Date de Naissance'].dt.year - 
        ((current_date.month < df_unique['Date de Naissance'].dt.month) | # v√©rifie si le mois actuel est avant le mois de naissance. Si c'est vrai, cela signifie que la personne n'a pas encore eu son anniversaire cette ann√©e.
         ((current_date.month == df_unique['Date de Naissance'].dt.month) & (current_date.day < df['Date de Naissance'].dt.day)))).tolist() # v√©rifie si c'est le m√™me mois mais avant le jour d'anniversaire. Si c'est vrai, cela signifie que la personne n'a pas encore eu son anniversaire ce mois-ci.

# Cr√©ation d'un DataFrame temporaire pour calculer l'√©cart-type des √¢ges
ages_df = pd.DataFrame(liste_ages, columns=['Ages'])

# Calculer l'√©cart-type des √¢ges
ect = ages_df['Ages'].std()

print("Ecart-type des √¢ges:", ect)
```

Nous pouvons remarquer que l'age de l'√©cart-type par rapport a la moyenne qui est de 61 ans est asser dissip√©e, cela veut dire que la plus part des ages se situe a 78ans (61+17) et 44ans (61-17)

# Sch√©ma repr√©sentant la moyenne d'age et l'√©cart-type sur le dataset de MUTEST

```python
# Cr√©er un histogramme des √¢ges
plt.figure(figsize=(10, 6))
plt.hist(liste_ages, bins=10, edgecolor='black')
plt.axvline(moyenne, color='red', linestyle='dashed', linewidth=1)
plt.axvline(moyenne+16, color='green', linestyle = 'dashed', linewidth=1)
plt.axvline(moyenne-16, color='green', linestyle = 'dashed', linewidth=1)
plt.title('Distribution des √¢ges')
plt.xlabel('√Çge')
plt.ylabel('Nombre de personnes')
plt.legend(['Moyenne d\'√¢ge', '√©cart-type'])
plt.show()
```

# Sch√©ma qui repr√©sente le nombre de personne en fonction de l'ann√©e

```python
import seaborn as sns
```

```python
# Cr√©ation d'un DataFrame pour les donn√©es √† visualiser
data = {
    'Ann√©e': [2022, 2023, 2024],
    'Nombre de personnes en attente': [nombre_en_attente_2022, nombre_en_attente_2023, nombre_en_attente_2024]
}
df_plot = pd.DataFrame(data)

# Cr√©er le diagramme en b√¢ton avec Seaborn
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Ann√©e', y='Nombre de personnes en attente', data=df_plot, palette='viridis', estimator="sum")
ax.bar_label(ax.containers[0], fontsize=10)
ax.bar_label(ax.containers[1], fontsize=10)
ax.bar_label(ax.containers[2], fontsize=10)

plt.title('Nombre de personnes en attente par ann√©e')
plt.xlabel('Ann√©e')
plt.ylabel('Nombre de personnes en attente')
plt.show()
```
